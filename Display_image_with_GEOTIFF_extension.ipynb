```python
import rasterio
import matplotlib.pyplot as plt
# تحديد مسار الصورة GeoTIFF
file_path = '/content/drive/MyDrive/Sentinel2_image.tif'  # استبدل هذا المسار بمسار ملف GeoTIFF الخاص بك

# فتح ملف GeoTIFF باستخدام Rasterio
with rasterio.open(file_path) as src:
    # قراءة القناة الأولى من الصورة (أو يمكنك قراءة أكثر من قناة)
    band1 = src.read(1)  # قراءة القناة الأولى

    # عرض معلومات عن الصورة (الإسقاط الجغرافي، حجم الصورة، إلخ)
    print(src.profile)

    # عرض الصورة باستخدام Matplotlib
    plt.imshow(band1, cmap='gray')  # عرض القناة الأولى بتدرجات اللون الرمادي
    plt.title('GeoTIFF Image')
    plt.colorbar()
    plt.show()
```

```python
# تحميل الصورة بامتداد GeoTIFF
file_path = '/content/drive/MyDrive/Sentinel2_image.tif'  # استبدل هذا المسار بمسار ملف GeoTIFF الخاص بك

# فتح ملف GeoTIFF باستخدام Rasterio
with rasterio.open(file_path) as src:
    # قراءة البيانات من القنوات
    image = src.read([1, 2, 3])  # هنا نقرأ القنوات الحمراء والخضراء والزرقاء (إذا كانت متاحة)

    # عرض الصورة باستخدام Matplotlib
    plt.imshow(image.transpose(1, 2, 0))  # تغيير ترتيب الأبعاد لعرض الصورة
    plt.title('GeoTIFF Image')
    plt.axis('off')  # إيقاف المحاور للحصول على عرض أوضح
    plt.show()
```

```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt
file_path = '/content/drive/MyDrive/Sentinel2_image.tif'  # استبدل هذا المسار بمسار ملف GeoTIFF الخاص بك
# فتح ملف GeoTIFF
with rasterio.open(file_path) as src:
    image = src.read([1, 2, 3])  # قراءة قنوات RGB

    # تطبيع القيم إلى النطاق [0, 1] إذا كانت القيم أكبر
    image = image.astype(np.float32)
    image /= image.max()  # تقسيم القيم على الحد الأقصى لجعلها في النطاق [0, 1]

    # عرض الصورة
    plt.imshow(image.transpose(1, 2, 0))  # ترتيب الأبعاد (العرض، الطول، القنوات)
    plt.title('Normalized GeoTIFF Image')
    plt.show()
```
